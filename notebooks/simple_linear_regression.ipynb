# Simple Linear Regression

## Example 1: Ice Cream Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../data/ice_cream_data.csv')
df.head(n=3)
```

```python
df.describe()
```

```python
sns.jointplot(x='Temperature', y='Revenue', data=df);
```

```python
sns.pairplot(df);
```

```python
sns.lmplot(x='Temperature', y='Revenue', data=df);
```

```python
from sklearn.model_selection import train_test_split

X = df[['Temperature']]
y = df['Revenue']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression().fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, c='gray', label='Training Set');
plt.plot(X_train, lm.predict(X_train), c='red', lw=2., label='Trained Model');
plt.xlabel('Feature (Temp)');
plt.ylabel('Value ($)');
plt.legend();
```

```python
plt.scatter(X_test, y_test, c='blue', label='Testing Set');
plt.plot(X_test, lm.predict(X_test), c='red', lw=2., label='Trained Model');
plt.xlabel('Feature (Temp)');
plt.ylabel('Value ($)');
plt.legend();
```

```python
print('Slope: {}\t Intercept: {}'.format(lm.coef_, lm.intercept_))
```

```python
from sklearn.metrics import r2_score

print('R^2: {:.2f}'.format(r2_score(y, lm.predict(X))))
```

## Example 2: Fuel Economy Data

```python
df = pd.read_csv('../data/fuel_economy.csv')
df.head(4)
```

```python
sns.jointplot(x='Horse Power', y='Fuel Economy (MPG)', data=df);
```

```python
sns.pairplot(df);
```

```python
sns.lmplot(x='Horse Power', y='Fuel Economy (MPG)', data=df);
```

```python
X = df[['Horse Power']]
y = df['Fuel Economy (MPG)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
lm = LinearRegression().fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, c='gray', label='Training Set');
plt.plot(X_train, lm.predict(X_train), c='red', lw=2., label='Trained Model');
plt.xlabel('Feature (HP)');
plt.ylabel('Value (MPG)');
plt.legend();
```

```python
plt.scatter(X_test, y_test, c='blue', label='Testing Set');
plt.plot(X_test, lm.predict(X_test), c='red', lw=2., label='Trained Model');
plt.xlabel('Feature (HP)');
plt.ylabel('Value (MPG)');
plt.legend();
```

```python
print('Slope: {}\t Intercept: {}'.format(lm.coef_, lm.intercept_))
```

```python
print('R^2: {:.2f}'.format(r2_score(y, lm.predict(X))))
```
