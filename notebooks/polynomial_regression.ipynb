# Polynomial Regression

## Example 1: Salary Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("../data/employee_salary.csv")
df.head(n=3)
```

```python
df.describe()
```

```python
sns.jointplot(x="Years of Experience", y="Salary", data=df);
```

```python
sns.pairplot(df);
```

```python
sns.lmplot(x="Years of Experience", y="Salary", data=df);
```

```python
from sklearn.model_selection import train_test_split

X = df[["Years of Experience"]]
y = df["Salary"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression(fit_intercept=False).fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, c="gray", label="Training Set")
plt.plot(X_train, lm.predict(X_train), c="red", lw=2.0, label="Trained Model")
plt.xlabel("Feature (Exp)")
plt.ylabel("Value ($)")
plt.legend();
```

```python
plt.scatter(X_test, y_test, c="blue", label="Testing Set")
plt.plot(X_test, lm.predict(X_test), c="red", lw=2.0, label="Trained Model")
plt.xlabel("Feature (Exp)")
plt.ylabel("Value ($)")
plt.legend();
```

```python
print("Slope: {}\t Intercept: {}".format(lm.coef_, lm.intercept_))
```

```python
from sklearn.metrics import r2_score

print("R^2: {:.2f}".format(r2_score(y, lm.predict(X))))
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2).fit(X_train)
X_poly_train = poly.transform(X_train)
```

```python
model = LinearRegression().fit(X_poly_train, y_train)
```

```python
print("Slope: {}\t Intercept: {}".format(model.coef_, model.intercept_))
```

```python
plt.scatter(X_train, y_train, c="gray", label="Training Set")
plt.plot(
    np.sort(X_train, axis=0),
    model.predict(np.sort(X_poly_train, axis=0)),
    c="red",
    label="Trained Model",
)
plt.xlabel("Feature (Exp)")
plt.ylabel("Value ($)")
plt.legend();
```

```python
X_poly_test = poly.transform(X_test)

plt.scatter(X_test, y_test, c="blue", label="Testing Set")
plt.plot(
    np.sort(X_test, axis=0),
    model.predict(np.sort(X_poly_test, axis=0)),
    c="red",
    label="Trained Model",
)
plt.xlabel("Feature (Exp)")
plt.ylabel("Value ($)")
plt.legend();
```

```python
poly = PolynomialFeatures(degree=5).fit(X_train)
X_poly_train = poly.transform(X_train)
X_poly_test = poly.transform(X_test)

model = LinearRegression().fit(X_poly_train, y_train)
print("Slope: {}\t Intercept: {}".format(model.coef_, model.intercept_))
```

```python
plt.scatter(X_train, y_train, c="gray", label="Training Set")
plt.plot(
    np.sort(X_train, axis=0),
    model.predict(np.sort(X_poly_train, axis=0)),
    c="red",
    label="Trained Model",
)
plt.xlabel("Feature (Exp)")
plt.ylabel("Value ($)")
plt.legend();
```

```python
plt.scatter(X_test, y_test, c="blue", label="Testing Set")
plt.plot(
    np.sort(X_test, axis=0),
    model.predict(np.sort(X_poly_test, axis=0)),
    c="red",
    label="Trained Model",
)
plt.xlabel("Feature (Exp)")
plt.ylabel("Value ($)")
plt.legend();
```

## Example 2: Economy Data

```python
df = pd.read_csv("../data/economies_of_scale.csv")
df.head(4)
```

```python
sns.jointplot(x="Number of Units", y="Manufacturing Cost", data=df);
```

```python
plt.scatter(X_test, y_test, c="blue", label="Testing Set")
plt.plot(
    np.sort(X_test, axis=0),
    model.predict(np.sort(X_poly_test, axis=0)),
    c="red",
    label="Trained Model",
)
plt.xlabel("Feature (#Units)")
plt.ylabel("Value ($)")
plt.legend();
```
